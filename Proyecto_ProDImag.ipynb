```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from matplotlib import colormaps
from google.colab.patches import cv2_imshow
import cv2
```

```python
#@title Importamos la imagen con la que deseamos trabajar :)

image_path = '/content/flores rosas.jpg'
img = mpimg.imread(image_path)

imgplot = plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
#@title Monitos 🐒

img_gray = np.dot(img[...,:3], [0.2989, 0.5870, 0.1140]) #a escala de grises

img_daltonismo_verdoso = img.copy()   #un tipo de daltonismo en monitos
#img_daltonismo_verdoso[:,:,0] = img[:,:,0]
img_daltonismo_verdoso[:,:,1] = img[:,:,0]

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(img_gray, cmap='gray')
axs[0].axis('off')

axs[1].imshow(img)
axs[1].axis('off')

axs[2].imshow(img_daltonismo_verdoso)
axs[2].axis('off')

plt.tight_layout()
plt.show()
```

Los monos del Viejo Mundo (África, Asia y Europa) ven el ídem de una forma bastante similar a los humanos:tricromático, azul, verde y rojo, pero los monos americanos registran hasta seis tipos de daltonismo diferente. Al igual que sucede en los humanos, el daltonismo es más frecuente en machos que en hembras.

```python
#@title Raton 🐁

#esta ya es una funcion para desenfocar, tramposilla yop
imagen_desenfocada = cv2.GaussianBlur(img, (31, 31), 0)

cv2_imshow(imagen_desenfocada)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Los ratones tienen una visión extraña. Pueden detectar los minimos movimientos de lo que está lejos de ellos, sin embargo, lo aprecian de manera desenfocada.
Por otro lado, pueden observar de manera más nitida lo que esta cerca a ellos, pero no detectan el movimiento, es decir, si estás inmovil frente a ellos, pueden no percatarse de tu existencia :)

```python
#@title Gatos 😺
rojo = img[:, :, 0]
verde = img[:, :, 1]
azul = img[:, :, 2]
amarillo = rojo * 2

imagen_gatos = np.stack([azul, verde, amarillo], axis=2) #aplico las escalas

plt.imshow(imagen_gatos)
plt.axis('off')
plt.show()
```

Los gatos aprecian las escalas de color azul, verde y amarillo, no siendo capaces de detectar los colores muy saturados o muy brillantes.

Abejas y plantas han coevolucionado durante millones de años, de modo que estas han adaptado los colores de sus pétalos para conseguir que los insectos las perciban fácilmente.

```python
#@title Caballo 🐎

height, width, _ = img.shape #sacar los valores de la imagen para poder generalizarlo

stripe_width = int(width * 0.1)

#aqui nada mas calcule las posiciones de la franja jaja
left_stripe_start = (width // 2) - (stripe_width // 2)
right_stripe_end = (width // 2) + (stripe_width // 2)

img_copy = img.copy()
img_copy[:, left_stripe_start:right_stripe_end] = 0

plt.imshow(img_copy)
plt.axis('off')
plt.show()
```

Los caballos tienen una increíble amplitud de miras, si exceptuamos lo que tienen delante de sus narices. Literalmente no pueden ver lo que está situado entre sus ojos, debido a su visión binocular. Eso explica que suelan mirar al suelo cuando caminan.

```python
#@title  Serpiente 🐍

gray_img = np.dot(img[...,:3], [0.2989, 0.5870, 0.1140])  #convertir la imagen a escala de grises
gray_img = gray_img / gray_img.max() #normalizar la imagen

colored_img = plt.cm.plasma(gray_img)#mapa de colores

plt.imshow(colored_img)
plt.axis('off')
plt.show()
```

 Las serpientes detectan tanto la luz visible (como lo hacemos nosotros) como la radiación infrarroja de una manera que nos es imposible imaginar. Los expertos estiman que este sofisticado sistema puede detectar cambios de temperatura de tan solo 0.002 grados centígrados.

```python
#@title Tiburon 🦈

imagen_bn = np.dot(img[..., :3], [0.2989, 0.587, 0.114])

plt.imshow(imagen_bn, cmap='gray')
plt.axis('off')
plt.show()
```

 Los tiburones no detectan los colores. Además, el ojo del tiburón está diseñado para absorber la máxima cantidad de luz posible, para poder ver en aguas turbulentas. Sin embargo, y aunque puede ver objetos a 20 metros de distancia, su visión no es tan fina como la nuestra.

```python
#@title Vaca 🐄
gray_img = np.dot(img[...,:3], [0.2989, 0.5870, 0.1140])
gray_img = gray_img / gray_img.max()

colored_img = plt.cm.hot(gray_img)

plt.imshow(colored_img)
plt.axis('off')
plt.show()
```

Un estudio revela que los bovinos pueden ver con mayor facilidad colores de longitud de onda larga o intermedia, de 700 a 550 nanómetros. En orden descendente, naranja>rojo>amarillo-verde>amarillo.

Estos animales tienen bastante dificultad para distinguir los colores de longitud de onda corta, de 500 a 400 nanómetros como el azul, gris y algunos verdes. El color que más se les dificulta ver es el violeta.
